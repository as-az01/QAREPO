```python
import re

# List of common titles to remove
titles = ["Mr", "Mrs", "Ms", "Dr", "Prof", "Sir", "Madam", "Lord", "Lady"]

# Compile a regex pattern to match titles followed by optional period and space
pattern = re.compile(r'\b(?:' + '|'.join(titles) + r')\.?\s+', re.IGNORECASE)

def clean_text(text):
    """
    Removes titles and converts the text to lowercase.
    """
    # Remove titles using regex
    text_without_titles = pattern.sub('', text)
    # Convert the result to lowercase
    return text_without_titles.lower()

# Example usage
if __name__ == "__main__":
    sample_texts = df['name']
    
    for text in sample_texts:
        cleaned = clean_text(text)
        print(f"Original: {text} --> Cleaned: {cleaned}")
```
